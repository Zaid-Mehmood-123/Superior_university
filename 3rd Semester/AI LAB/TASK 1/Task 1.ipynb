```python
import random

class hangman:
    def __init__(self, random_word, attempts=6):
        self.random_word = random_word
        self.attempts = attempts
        self.selected_word = random.choice(self.random_word)
        self.current_progress = ['_' for _ in range(len(self.selected_word))]

        self.hangman_stages = [
            "   -----",
            "   |   |",
            "   |   O",
            "   |  /|\\",
            "   |  / \\",
            "   |"
        ]

        self.is_game_over = False
    
    def play_game(self):
        while not self.is_game_over:
            player_guess = input("Guess a letter: ").lower()
            self.update_display(player_guess)
            self.check_guess(player_guess)
            self.check_game_status()
    
    def update_display(self, player_guess):
        for index in range(len(self.selected_word)):
            if self.selected_word[index] == player_guess:
                self.current_progress[index] = player_guess
        print("Current Progress: ", ' '.join(self.current_progress))
    
    def check_guess(self, guessed_letter):
        if guessed_letter not in self.selected_word:
            self.attempts -= 1
            stage_line_index = len(self.hangman_stages) - self.attempts
            for i in range(stage_line_index):
                print(self.hangman_stages[i])
            if self.attempts == 0:
                self.is_game_over = True
                print("loserrrrr You've lost the game!")
                print("GAME OVER!!!!!!")
    
    def check_game_status(self):
        if '_' not in self.current_progress:
            self.is_game_over = True
            print("Congratulations, you've won the game!")

word_pool = ['rasikh', 'zaid', 'robot']
game_instance = hangman(word_pool)
game_instance.play_game()
```
